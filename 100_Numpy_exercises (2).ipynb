#### 19. Create a 8x8 matrix and fill it with a checkerboard pattern (★☆☆)

````python
# Create an 8x8 checkerboard pattern using ★ and ☆
size = 8
for row in range(size):
    line = ''
    for col in range(size):
        if (row + col) % 2 == 0:
            line += '★ '
        else:
            line += '☆ '
    print(line.strip())
````
