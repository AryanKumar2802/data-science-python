```python
import pandas as pd

# Assuming your DataFrame is named 'df' and you loaded it correctly

# Select only numeric columns for finding the minimum value
numeric_df = df.select_dtypes(include=['number'])

# Find the minimum value in the numeric columns
min_value = numeric_df.min().min()  # First min finds min per column, second finds min overall

# Find the row containing the minimum value (consider all columns)
row_with_min_value = df[df.eq(min_value).any(axis=1)]

# Display the row
print(row_with_min_value)
```
